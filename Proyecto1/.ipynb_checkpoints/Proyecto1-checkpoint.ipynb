### _Carga y separación de los datos_

```python
import numpy as np

#Carga del dataset
dataset = np.load("proyecto_training_data.npy")

filas_count = dataset.shape[0]
print(f"Número total de filas: {filas_count}")
punto_corte = int(filas_count * 0.8)
set_entrenamiento = dataset[:punto_corte]
set_validacion = dataset[punto_corte:]
print(f"Filas para entreamiento: {set_entrenamiento.shape[0]}")
print(f"Filas para validación y pruebas: {set_validacion.shape[0]}")
```

### _Análisis exploratorio de los datos_

Construcción del DataFrame de Pandas

```python
import pandas as pd

#Nombres de los campos, según el archivo 'Data fields.txt'
data_fields = [
    'SalePrice',
    'OverallQual',
    '1stFlrSF',
    'TotRmsAbvGrd',
    'YearBuilt',
    'LotFrontage'
]

df_datos = pd.DataFrame(dataset, columns=data_fields)
df_datos.head()
```

Principales datos estadísticos e histograma para cada variable

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in df_datos:
    
    #obtención de datos estadísticos
    df_describe = df_datos[col].describe()
    print(col)
    print("Media: ", df_describe['mean'])
    minimo = df_describe['min']
    maximo = df_describe['max']
    rango = maximo - minimo
    print("Valor mínimo: ", minimo)
    print("Valor máximo:", maximo)
    print("Rango:", rango)
    print("Desviación estándar:", df_describe['std'])
    
    #Despliegue del histograma con seaborn
    sns.histplot(df_datos[col])
    plt.show()
    
    print('-'*80)
    print()
```
